#### Dealing with missing data

Its Given that some data is missing in the data set description

 Attribute #:   Number of instances missing a value:
   2.             41    //normalized loses
   6.             2      //num-of-doors
   19.            4     //bore
   20.            4      //stroke
   22.            2     //horsepower
   23.            2      //peak-rpm
   26.            4      //price
   
   
   Deal with these missing values: Depends on the columns the missing values are at
   In some training data rows the price is missing
   We want to train the model to predict the price using given data
   Now when price is missing the whole row( training data row) is useless
   So we drop it

```python
df.head(2)
```

```python
# Replace ? with NAN
df=df.replace('?',np.NaN)
```

Below we get all the rows which have atleast one missing cell
We do this to understand how we can deal with the missing values

```python
df.loc[df.isnull().any(axis=1)]
```

num of doors of sedan car are missing
Usually it has 4 doors
So we replace the num of doors by 4 instead of NAN.

```python
df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

The stroke and bore values of mazda style automobile which runs on gas is missing
We dont have any other rows with these details for reference
So we will replace them with mean


```python
avg_bore=df['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)

#Replace NaN by mean value
df["bore"].replace(np.nan, avg_bore, inplace=True)

avg_stroke=df['stroke'].astype('float').mean(axis=0)
print("Average of stroke:", avg_stroke)

#Replace NaN by mean value
df["stroke"].replace(np.nan, avg_stroke, inplace=True)
```

```python
avg_horsepower=df['horsepower'].astype('float').mean(axis=0)
print("Average of horsepower:", avg_horsepower)

#Replace NaN by mean value
df["horsepower"].replace(np.nan, avg_bore, inplace=True)

avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average of peak_rpm:", avg_peakrpm)

#Replace NaN by mean value
df["peak-rpm"].replace(np.nan, avg_peakrpm, inplace=True)
```

```python
# Calculate the average of the column
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)

#Replace "NaN" by mean value in "normalized-losses" column
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
# Calculate the average of the column
avg_price = df["price"].astype("float").mean(axis=0)
print("Average of prices:", avg_price)

#Replace "NaN" by mean value in "normalized-losses" column
df["price"].replace(np.nan, avg_price, inplace=True)
```
